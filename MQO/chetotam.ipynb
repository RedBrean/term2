```python
from pathlib import Path
import re

import numpy as np
from PIL import Image

# ==========================================
# НАСТРОЙКИ
# ==========================================

input_dir = Path(r"C:\Users\RedBrean\Documents\MIPT\term2\mqo\19_06_2026")
output_dir = input_dir / "processed"

output_dir.mkdir(exist_ok=True)

# ==========================================
# ФУНКЦИИ
# ==========================================

def load_image(fname):
    """
    Загружает PNG как float-массив.
    """
    return np.array(Image.open(fname), dtype=np.float64)


def average_images(files):
    """
    Попиксельное усреднение списка изображений.
    """
    stack = np.stack([load_image(f) for f in files])
    return stack.mean(axis=0)


def save_image(arr, fname):
    """
    Сохраняет массив как PNG.
    """
    arr = np.clip(arr, 0, 65535)

    if arr.max() <= 255:
        arr = arr.astype(np.uint8)
    else:
        arr = arr.astype(np.uint16)

    Image.fromarray(arr).save(fname)


# ==========================================
# РАЗБОР ИМЁН
# ==========================================

groups = {}

for file in input_dir.glob("*.png"):

    name = file.stem

    if "_NOISE_" in name:
        key = name.split("_NOISE_")[0]
        kind = "noise"
    else:
        m = re.match(r"(.+?)_\d{4}-\d{2}-\d{2}T", name)

        if m:
            key = m.group(1)
        else:
            key = "_".join(name.split("_")[:2])

        kind = "signal"

    groups.setdefault(key, {"signal": [], "noise": []})
    groups[key][kind].append(file)

# ==========================================
# ОБРАБОТКА
# ==========================================

for key, data in sorted(groups.items()):

    signal_files = data["signal"]
    noise_files = data["noise"]

    if len(signal_files) == 0:
        continue

    print(f"\n{key}")
    print(f"  signal: {len(signal_files)}")
    print(f"  noise : {len(noise_files)}")

    signal_avg = average_images(signal_files)

    if len(noise_files) > 0:
        noise_avg = average_images(noise_files)
        result = signal_avg - noise_avg
    else:
        result = signal_avg

    result[result < 0] = 0

    save_image(
        result,
        output_dir / f"{key}_AVG_MINUS_NOISE.png"
    )

print("\nГотово.")
print("Результаты лежат в:")
print(output_dir)
```
