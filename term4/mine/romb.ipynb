```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt


# Вычисление фазы
def delta_phi(theta, n):
    kappa = np.sqrt(np.sin(theta)**2 - 1 / n**2)
    term1 = np.arctan(kappa / np.cos(theta))
    term2 = np.arctan(n**2 * kappa / np.cos(theta))
    return -2 * (term1 - term2)
```

```python
# Углы от критического до 90 градусов
theta = np.linspace(0, np.pi / 2 - 0.01, 500)


# Построение графика
plt.figure(figsize=(8, 5))
plt.plot(np.degrees(theta), delta_phi(theta,2.41), label=r'$\Delta \varphi$')
plt.axhline(np.pi / 2, color='red', linestyle='--', label=r'$\frac{\pi}{2}$')
plt.xlabel(r'Угол падения $\theta$ (градусы)')
plt.ylabel(r'Фазовый сдвиг $\Delta \varphi$ (рад)')
plt.title('Фазовый сдвиг при полном внутреннем отражении (n = 1.5)')
plt.grid(True)
plt.savefig("graph.png", dpi=300)
plt.legend()
plt.tight_layout()
plt.show()

# Define the equation to solve
def equation_to_solve(theta):
    return delta_phi(theta, 2.41) - np.pi / 4

# Initial guess for theta (in radians)
initial_guess = np.radians(61)  # 45 degrees

# Solve the equation
# solution = fsolve(equation_to_solve, initial_guess)

# Print the solution
# print(f"The solution for delta_phi(theta, {1.7}) = π/4 is theta = {np.degrees(solution[0]):.4f} degrees")
```
