```python
import numpy as np
import matplotlib.pyplot as plt

# Данные
arr1 = np.array([0.203, 0.219, 0.247, 0.249, 0.274, 0.29, 0.292, 0.308, 0.315])
arr2 = np.array([-0.202, -0.218, -0.246, -0.248, -0.273, -0.289, -0.29, -0.306, -0.312])
arr3 = np.array([405.18, 437.63, 494.61, 497.13, 548.48, 579.95, 583.78, 615.26, 630.65])
arr4 = np.array([403.58, 435.45, 491.48, 495.94, 546.34, 577.24, 579.73, 612.95, 623.98])


# Подписи осей
plt.xlabel("$\lambda\\text{нм}$")
plt.ylabel("$\sin(\\varphi)$")



# Построение графиков
plt.scatter(arr3, arr1, color='blue', label='$m=+1$')
plt.scatter(arr4, -1*arr2, color='red', label='$m=-1$')

# Заголовок и легенда
plt.legend()

# Сетка и отображение
plt.grid(True)
plt.savefig("graph", dpi=300)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Данные
arr1 = np.array([1, -1, 2, -2, 3, -3])
arr2 = np.array([9.5, -5.7, 11.4, -6.7, 9.5, -10.9])
arr3 = arr1*5



# Построение графиков
plt.scatter(arr1, arr2, color='blue', label='$Эксперимент$')
plt.scatter(arr1, arr3, color='red', label='$Теория$')

# Подписи осей
plt.xlabel("$m$", fontsize=14)
plt.ylabel("$D, 10^5m$", fontsize=14)

# Заголовок и легенда
plt.legend()

# Сетка и отображение
plt.grid(True)
plt.savefig("graph2", dpi=300)
plt.show()
```
