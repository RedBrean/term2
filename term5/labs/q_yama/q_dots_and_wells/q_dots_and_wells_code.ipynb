```python
import pandas as pd
import matplotlib.pyplot as plt

files_list = ['a581','c262','c263','no_name']
descriptions_list = [r'q-well, sample a581',r'q-well, sample c262',r'q-well, sample c263',r'q-well, sample without label']
intervals_list = [[810,870],[800,840],[730,790],[730,790]]
df = [0,1,2,3]
# Read and process the data
for i in [0,1,2,3]:
    df[i] = pd.read_csv(f'q_dots_and_wells/{files_list[i]}.csv', sep=' ')
    df[i] = df[i][df[i].columns[:2]][1:]
    df[i].columns = ['x', 'y']
    df[i] = df[i].apply(pd.to_numeric, errors='coerce')
    df[i] = df[i][df[i].iloc[:, 0] > intervals_list[i][0]]
    df[i] = df[i][df[i].iloc[:, 0] < intervals_list[i][1]]
    plt.plot(df[i].iloc[:, 0], df[i].iloc[:, 1])
    plt.grid()
    plt.title(descriptions_list[i])
    plt.xlabel(r'$\lambda,$ nm')
    plt.ylabel(r'$U?$')
    plt.savefig(f'q_dots_and_wells/q_well_{files_list[i]}.png')

    plt.show()
```
