```python
import numpy as np
from scipy.optimize import least_squares

sphere_points = np.array([[1542,1462,2.940],[1399,1645,2.881],[1686,1306,2.814],[1427,1297,2.838],[1642,1658,2.894]])
def sphere_res(params, pts):
    x0,y0,z0,r = params
    return np.sqrt((pts[:,0]-x0)**2 + (pts[:,1]-y0)**2 + (pts[:,2]-z0)**2) - r

centroid = sphere_points.mean(axis=0)
r0 = np.mean(np.sqrt(((sphere_points - centroid)**2).sum(axis=1)))
p0 = [centroid[0], centroid[1], centroid[2], r0]

res = least_squares(sphere_res, p0, args=(sphere_points,))
x0,y0,z0,r = res.x

print(f"center = ({x0:.6f}, {y0:.6f}, {z0:.6f}), radius = {r:.6f}")
print("rmse =", np.sqrt(np.mean(res.fun**2)))
```
