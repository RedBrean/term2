```python
import numpy as np

def generate_random_vector(cov, mean=None):
    A = np.linalg.cholesky(cov)
    n = cov.shape[0]
    z = np.random.randn(n)
    if mean is None:
        mean = np.zeros(n)
    x = mean + A @ z
    return x
cov = np.array([[1.0, 0.5], [0.5, 1.0]])
mean = np.array([0, 0])

x = generate_random_vector(cov, mean)
print(x)
```
