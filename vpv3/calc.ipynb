```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
i = 0 + 1j

```

```python
def amp(f, C1, C2, R1, R2):
    w = f*6.28
    return 10*np.log(abs((1/(1 - w**2*C1*C2*R1*R2+ i*w*C2*(R1+R2)))))

def fas(f, C1, C2, R1, R2):
    w = f*6.28
    return np.angle((1/(1 - w**2*C1*C2*R1*R2+ i*w*C2*(R1+R2))))

C1 = 2 * 10**-9
C2 = 0.3 * 10**-9
R1 = 50 * 10**3
R2 = 100 * 10**3 - R1
fs = np.linspace(30, 10000, 1000)
plt.plot(fs, [amp(f, C1, C2, R1, R2) for f in fs])
plt.figure()
plt.plot(fs, [fas(f, C1, C2, R1, R2) for f in fs])
```
