```python
def P(n,m):
    if m <= 0:
        return 0.0
    if m == 1:
        return (n%12 == 0)*1.0
    if n%12 == 0:
        return (m == 1)*1.0
    #print("Calculating P({}, {}), result is {}".format(n, m, P((n-1)%12,m-1)*0.5 + P((12-n)%12,m-1) * (0.5)**(2*n)))
    return P((n-1)%12,m-1)*0.5 + P((12-n)%12,m-1) * (0.5)**(2*n)

print(P(2,3))
     
```
